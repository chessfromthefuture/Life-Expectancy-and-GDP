```python
import pandas as pd
```

```python
data = pd.read_csv('all_data.csv')
data.head()
```

```python
data.columns
```

```python
print(data.info())
print(data.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Line plot of Life Expectancy over Years for each Country
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.title('Trend of Life Expectancy Over Years by Country')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (Years)')
plt.legend(title='Country')
plt.show()
```

```python
# Scatter plot with GDP vs. Life Expectancy
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='GDP', y='Life expectancy at birth (years)', hue='Country', palette='viridis')
plt.title('GDP vs. Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy at Birth (Years)')
plt.show()
```

```python
# Boxplot of Life Expectancy for each Country
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Country', y='Life expectancy at birth (years)')
plt.title('Distribution of Life Expectancy by Country')
plt.xlabel('Country')
plt.ylabel('Life Expectancy at Birth (Years)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(data[['Life expectancy at birth (years)', 'GDP']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Variables')
plt.show()
```

```python
# Line plot of GDP over Years
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='GDP', hue='Country', marker='o')
plt.title('GDP Growth Over Years by Country')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(title='Country')
plt.show()
```

# has life expectancy increased

# Has GDP increased over time in the six nations?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='GDP', hue='Country', marker='o')
plt.title('GDP Trends Over Time by Country')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(title='Country')
plt.show()
```

# Is There a Correlation Between GDP and Life Expectancy?

```python
# Scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='GDP', y='Life expectancy at birth (years)', hue='Country', palette='viridis')
plt.title('Correlation Between GDP and Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy at Birth (Years)')
plt.show()

# Correlation coefficient
correlation = data[['GDP', 'Life expectancy at birth (years)']].corr().iloc[0, 1]
print(f"Correlation between GDP and Life Expectancy: {correlation:.2f}")
```

# 4. What Is the Average Life Expectancy in These Nations?
You can calculate and visualize the average life expectancy per country.

```python

# Average Life Expectancy per Country
avg_life_exp = data.groupby('Country')['Life expectancy at birth (years)'].mean()

print("Average Life Expectancy per Country:")
print(avg_life_exp)

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_life_exp.index, y=avg_life_exp.values, palette='pastel')
plt.title('Average Life Expectancy by Country')
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy (Years)')
plt.xticks(rotation=45)
plt.show()
```

# What Is the Distribution of Life Expectancy?
A histogram or KDE plot can show the distribution of life expectancy.


```python


plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Life expectancy at birth (years)', kde=True, bins=20, color='skyblue')
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy at Birth (Years)')
plt.ylabel('Frequency')
plt.show()
```
